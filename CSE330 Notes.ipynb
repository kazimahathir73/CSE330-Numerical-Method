**Bisection Method**

```python
''' sp.log(x , 10) || sp.sin(x) || sp.cos(x) || sp.tan(x) || sp.cot(x) || sp.sec(x) || sp.csc(x) || sp.exp(x)'''
''' changes the value '''
def f(x):
    x4_coef = 0
    x3_coef = 2
    x2_coef = 0
    x1_coef = -2
    x0_coef = -5
    return x4_coef*x**4 + x3_coef*x**3 + x2_coef*x**2 + x1_coef*x + x0_coef

a = 1.2
b = 1.8
e_value = -8
e = 10**e_value
max_iterations = 15

''' changes the value '''

# Populate the following lists in each iteration
list_a = []
list_b = []
list_m = []
list_f_a = []
list_f_b = []
list_f_m = []
interval = []

iteration = 0

while abs(b - a) > e and iteration < max_iterations:
    m = (a + b) / 2
    list_a.append(round(a,5))
    list_b.append(round(b,5))
    list_m.append(round(m,5))
    list_f_a.append(round(f(a),5))
    list_f_b.append(round(f(b),5))
    list_f_m.append(round(f(m),5))
    interval.append((round(a, 3), round(b, 3)))

    if f(a) * f(m) < 0:
        b = m
    elif f(b) * f(m) < 0:
        a = m
    elif f(m) == 0:
        break
    iteration += 1

# Append the final iteration with the last interval
list_a.append(a)
list_b.append(b)
list_m.append(m)
list_f_a.append(f(a))
list_f_b.append(f(b))
list_f_m.append(f(m))
interval.append((round(a, 5), round(b, 5)))

print(pd.DataFrame({'a': list_a, 'm': list_m, 'b': list_b, 'f(a)': list_f_a, 'f(m)': list_f_m, 'f(b)': list_f_b, 'interval': interval}))
```

```python
x = sp.symbols('x')

''' sp.log(x , 10) || sp.sin(x) || sp.cos(x) || sp.tan(x) || sp.cot(x) || sp.sec(x) || sp.csc(x) || sp.exp(x)'''
''' changes the value '''
x4_coef = 0
x3_coef = 0
x2_coef = 1
x1_coef = -2 * sp.exp(-x)
x0_coef = sp.exp(-2 * x)
e_value = -3
x_value = 0
min_itr = 100
''' changes the value '''

f = x4_coef * x**4 + x3_coef * x**3 + x2_coef * x**2 + x1_coef * x + x0_coef
#f = sp.cos(2*x) - sp.sin(x)
f_prime = sp.diff(f, x)
e = 10**e_value
x_k = [x_value]
f_x_k = [float(f.subs(x, x_value).evalf())]
yes_no = ['Initial']
k = 0

while abs(f.subs(x, x_value)) > e and k < min_itr:
    f_value = float(f.subs(x, x_value).evalf())
    f_prime_value = float(f_prime.subs(x, x_value).evalf())

    if f_prime_value == 0:
        print("Derivative is zero, cannot proceed further.")
        break

    x_value_old = x_value
    x_value = x_value - (f_value / f_prime_value)
    x_k.append(x_value)
    f_x_k.append(float(f.subs(x, x_value).evalf()))
    yes_no.append('No')

    if k % 2 == 1:
        x_k_minus_1 = x_k[-3]
        x_k_k = x_k[-2]
        x_k_plus_1 = x_k[-1]
        delta = x_k_plus_1 - 2 * x_k_k + x_k_minus_1
        if delta != 0:
            x_k_aitken = x_k_minus_1 - ((x_k_k - x_k_minus_1)**2) / delta
            x_value = x_k_aitken
            x_k.append(x_value)
            f_x_k.append(float(f.subs(x, x_value).evalf()))
            yes_no.append('Yes (Aitken)')
        else:
            yes_no[-1] = 'No (Skip Aitken)'
    k += 1

if len(x_k) == len(f_x_k) and len(f_x_k) == len(yes_no):
    print(pd.DataFrame({'xk': x_k, 'f(xk)': f_x_k, 'Yes/No': yes_no}))
else:
    print("Error: Mismatch in array lengths.")
```

**QR Decomposition**

```python
def qr_decomposition(A):
    """
    Perform QR decomposition of matrix A using the Gram-Schmidt process.
    Args:
    A (numpy.ndarray): The input matrix

    Returns:
    Q (numpy.ndarray): The orthogonal matrix
    R (numpy.ndarray): The upper triangular matrix
    """
    m, n = A.shape
    Q = np.zeros((m, n))
    R = np.zeros((n, n))

    for j in range(n):
        # Compute the j-th column of Q
        q = A[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i].T, A[:, j])
            q = q - R[i, j] * Q[:, i]

        print(f"P{j+1} {q}")
        R[j, j] = np.linalg.norm(q)
        Q[:, j] = q / R[j, j]
        print(f"q{j+1} {Q[:, j]}")
        print()

    return Q, R

def solve_qr(Q, R, b):
    """
    Solve the linear system Ax = b using QR decomposition.
    Args:
    Q (numpy.ndarray): The orthogonal matrix from QR decomposition
    R (numpy.ndarray): The upper triangular matrix from QR decomposition
    b (numpy.ndarray): The right-hand side vector

    Returns:
    x (numpy.ndarray): The solution vector
    """
    # Compute Q^T b
    Qt_b = np.dot(Q.T, b)

    # Back substitution to solve R x = Q^T b
    x = np.linalg.solve(R, Qt_b)

    return Qt_b, x

# <<<<<<<<<<<<<<<<<<<<<<<<< change the part here >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>

A = np.array([[1,-3],
              [1, 0],
              [1, 6]])             # Example matrix
b = np.array([0,
              0,
              2])                  # Example vector

# <<<<<<<<<<<<<<<<<<<<<<<<< change the part here >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>


# Perform QR decomposition
Q, R = qr_decomposition(A)

# Print the results
print("Matrix Q:\n", Q)
print("\nMatrix R:\n", R)

# Solve Ax = b
Qt_b, x = solve_qr(Q, R, b)
print("\nQT*b:\n", Qt_b)
print("\nSolution x:\n", x)
```

**Tarpezium Rules**

```python
def trapezoidal_rule(func, a, b, n):
    h = (b - a) / n
    print(f"h: {h}")
    x_values = np.linspace(a, b, n + 1)

    integral = 0.5 * (func(a) + func(b))  # Start with the endpoints

    print("\nStep 0:")
    print(f"f({a}) = {func(a)}, f({b}) = {func(b)}")

    for i in range(1, n):
        x_i = x_values[i]
        integral += func(x_i)

        print(f"\nStep {i}:")
        print(f"f({x_i}) = {func(x_i)}")

    integral *= h

    print("\nFinal Result:")
    print(f"Integral from {a} to {b} using Trapezoidal Rule: {integral}")

    return integral

""" <<<<<<<<<<<<<<<<<<<<<<< Change the value >>>>>>>>>>>>>>>>>>>>>> """
def example_function(x):
    x4_coef = 0
    x3_coef = 0
    x2_coef = 1
    x1_coef = 1
    x0_coef = 6

    #f = x4_coef * x**4 + x3_coef * x**3 + x2_coef * x**2 + x1_coef * x + x0_coef
    #f = x**2 - sp.exp(-2*x) -x
    #f = 1/x - 0.5
    f= sp.exp(0.5*x) + sp.sin(x)
    return f

a = 0           # lower bound
b = 2           # upper bound
n = 4
result = trapezoidal_rule(example_function, a, b, n)
""" <<<<<<<<<<<<<<<<<<<<<<< Change the value >>>>>>>>>>>>>>>>>>>>>> """
```

**Simpson Rules**

```python
def simpsons_rule(func, a, b, n):
    h = (b - a) / n
    x_values = np.linspace(a, b, n + 1)

    integral = func(a) + func(b)  # Start with the endpoints

    print("\nStep 0:")
    print(f"f({a}) = {func(a)}, f({b}) = {func(b)}")

    for i in range(1, n):
        x_i = x_values[i]
        weight = 4 if i % 2 == 1 else 2
        integral += weight * func(x_i)

        print(f"\nStep {i}:")
        print(f"f({x_i}) = {func(x_i)} (Weight: {weight})")

    integral *= h / 3

    print("\nFinal Result:")
    print(f"Integral from {a} to {b} using Simpson's Rule: {integral}")

    return integral

""" <<<<<<<<<<<<<<<<<<<<<<< Change the value >>>>>>>>>>>>>>>>>>>>>> """
def example_function(x):
    x4_coef = 0
    x3_coef = 0
    x2_coef = 1
    x1_coef = 1
    x0_coef = 6

    f = x4_coef * x**4 + x3_coef * x**3 + x2_coef * x**2 + x1_coef * x + x0_coef
    #f = x**2 - sp.exp(-2*x) -x
    #f = 1/x - 0.5
    return f

a = 0           # lower bound
b = 2           # upper bound
n = 4
result = simpsons_rule(example_function, a, b, n)
""" <<<<<<<<<<<<<<<<<<<<<<< Change the value >>>>>>>>>>>>>>>>>>>>>> """
```
